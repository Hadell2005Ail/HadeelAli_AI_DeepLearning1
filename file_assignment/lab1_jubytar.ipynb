```python
import numpy as np
import matplotlib.pyplot as plt

#input
X=np.array([
    [20,10000,20],
      [30,20000,10],
        [40,130000,2]
])



#nural
def liner(x,w,b):
    z= np.dot(w,x)+b
    return z
w=np.array([0.05,0.001,0.4])
b=-2


z=liner(X,w,b)
print(z)

#sigmoud
def sigmoud(z):
    return 1/(1+np.exp(-z))

#probility
probility=sigmoud(z)
np.round(probility,3)
print(probility)

predicted_class = (probility>= 0.5).astype(int)

print("Actual result:   ", X.astype(int))
print("Predicted result:", predicted_class)




 
```

```python
age=X[:,0]
visit=X[:,2]
plt.figure(figsize=(8,5))
plt.scatter(age,visit, color="blue")
plt.plot(age,visit,color="blue",label="Line")
plt.xlabel("Age")
plt.ylabel("websitvisit")
plt.title("Rlation beetween age and website visits")
plt.grid(True)
plt.show()
```
